```python
data_dir = '/checkpoint/apratik/data_devfair0187/apartment_0/straightline/no_noise/1633991019'

num_traj = 3

import os
import matplotlib.pyplot as plt
import cv2
from IPython.display import Video
import json

def visualize(img_ids, traj):
    rgb_d = os.path.join(data_dir, str(traj), 'default', 'rgb_dbg')
    if not os.path.isdir(rgb_d):
        print(f'{rgb_d} does not exist')
        return
    print(f'visualizing {img_ids} ...')
    rows = int(len(img_ids)/5)
    fig, axs = plt.subplots(rows, 5, figsize=(30,5*rows), dpi=160)
    ctr = 0
    
    if rows > 1:
        for x in range(rows):
            for y in range(5):
                img_path = os.path.join(rgb_d, "{:05d}.jpg".format(img_ids[ctr]))
                axs[x,y].imshow(cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB))
                axs[x,y].set_title("{:05d}.jpg".format(img_ids[ctr]))
                ctr += 1
            
    elif rows == 1:
        for x in range(len(img_ids)):
            img_path = os.path.join(rgb_d, "{:05d}.jpg".format(img_ids[x]))
            axs[x].imshow(cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB))
            axs[x].set_title("{:05d}.jpg".format(img_ids[x]))
        
    plt.show()
    
job_dir = '/checkpoint/soumith/jobs/active_vision/pipeline/apartment_0/straightline/no_noise/12-11-2021/13:03:17'

for traj in range(10):
    traj_dir = os.path.join(job_dir, str(traj), 'default')
    if os.path.isdir(traj_dir):
        # get img ids chosen, visualize them for different gt_p pairs
        for gt in range(5, 30, 5):
            for p in range(0, 30, 5):
                pred_f = os.path.join(traj_dir, f'pred_label_gt{gt}p{p}')
                train_ids = os.path.join(pred_f, 'train_img_id.json')
                if os.path.isfile(train_ids):
                    with open(train_ids, 'r') as f:
                        d = json.load(f)
                        visualize(d['img_id'], traj)
                    break
                
        

        
#     if os.pth

# # for x in range(2, num_traj):
# x = 1
# traj_path = os.path.join(data_dir, str(x), 'default')
# print(f'traj_path {traj_path}')
# if os.path.isdir(traj_path):
#     # visualize the rgb_dbg dir
#     rgb = os.path.join(traj_path, 'rgb_dbg')
#     imgs = os.listdir(rgb)
#     n = len(imgs)
#     ctr = 0
#     rows = n/5
#     for x in range(int(rows)):
#         fig, axs = plt.subplots(1, 5, figsize=(30,8))
#         for y in range(5):
#             axs[y].imshow(cv2.cvtColor(cv2.imread(os.path.join(rgb, imgs[ctr])), cv2.COLOR_BGR2RGB))
#             axs[y].set_title(imgs[ctr])
#             ctr+=1
#         plt.show()
```
