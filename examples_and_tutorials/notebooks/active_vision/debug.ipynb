```python
import os
import matplotlib.pyplot as plt
import random
import cv2
from detectron2.utils.visualizer import Visualizer, ColorMode
from detectron2.data import DatasetCatalog, MetadataCatalog
from detectron2.data.datasets import register_coco_instances

from matplotlib.pyplot import figure

plt.rcParams["figure.figsize"] = (20,10)

import subprocess
from IPython.display import Video, HTML

def visualize_video(img_dir, save_dir, traj):
    out_d = os.path.join(os.getcwd(), f'videos/{save_dir}')
    if not os.path.exists(out_d):
        os.makedirs(out_d)
    out_f = os.path.join(out_d, f'{traj}.gif')
    if os.path.exists(out_f):
        os.remove(out_f)
    command = f"ffmpeg -framerate 50 -f image2 -i {img_dir}/%05d.jpg {out_f}"
#     command = f"ffmpeg -framerate 50 -i {img_dir}/%05d.jpg -c:v libx264 -vf 'fps=25,format=yuv420p' {out_f}"
    print(command)
    subprocess.call(command, shell=True)
    
#     Video(f"videos/{traj}.mp4")
#     Video("videos/hotel_0_curious.mp4", embed=True)

    
    
def visualize_coco(train_json, img_dir, pred_dir):
    DatasetCatalog.clear()
    MetadataCatalog.clear()
    register_coco_instances('hamilton', {}, train_json, img_dir)
    dataset_dicts = DatasetCatalog.get('hamilton')

    print(f'{pred_dir} len dataset {len(dataset_dicts)}')
        
    i = 0
    j = 0
    fig, axs = plt.subplots(1, 4)
    for d in random.sample(dataset_dicts, min(len(dataset_dicts), 4)):
        img = cv2.imread(d["file_name"])
        visualizer = Visualizer(img[:, :, ::-1], metadata=MetadataCatalog.get('hamilton'), scale=0.5)
        vis = visualizer.draw_dataset_dict(d)
        img = vis.get_image()
        
        axs[j].imshow(img)
        j += 1
        if j == 4:
            j = 0
            i += 1
#         plt.figure(figsize=(12,8))
#         plt.imshow(img)
    plt.show()

def examine_labels(pred_dir, img_dir, traj, save_dir):
    coco = os.path.join(pred_dir, 'coco_train.json')
    if os.path.isfile(coco):
        visualize_video(img_dir, save_dir, traj)
#         visualize_coco(coco, img_dir, pred_dir)
    
def debug(log, root_data_dir, save_dir, num_traj=20):
    for traj in range(num_traj):
        traj_root = os.path.join(log, str(traj))
        print(f'traj_root {traj_root}')
        if os.path.isdir(traj_root):
            img_dir = os.path.join(root_data_dir, str(traj), 'rgb')
            print(f'img_dir {img_dir}, num_images {len(os.listdir(img_dir))} ')
            for x in random.sample(os.listdir(traj_root), min(len(os.listdir(traj_root)), 2)):
                examine_labels(os.path.join(traj_root, x), img_dir, traj, save_dir)


def debug_vid(root_data_dir, save_dir, num_traj=20):
    for traj in range(num_traj):
        traj_root = os.path.join(root_data_dir, str(traj))
        print(f'traj_root {traj_root}')
        if os.path.isdir(traj_root):
            img_dir = os.path.join(traj_root, 'rgb')
            visualize_video(img_dir, save_dir, traj)
```
