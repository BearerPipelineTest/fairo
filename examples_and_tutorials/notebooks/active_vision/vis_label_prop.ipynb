```python
job_dir = '/checkpoint/apratik/jobs/active_vision/pipeline/instance_det/apartment_0/straightline/no_noise/11-11-2021/11:34:29'
data_dir = '/checkpoint/apratik/data_devfair0187/apartment_0/straightline/no_noise/instance_detection_ids_allinone'

# job_dir = '/checkpoint/apratik/jobs/active_vision/pipeline/instance_det/apartment_0/default/no_noise/10-11-2021/23:14:46'
# data_dir = '/checkpoint/apratik/data/data/apartment_0/default/no_noise/instance_detection_ids_allinone'
gt = 5
p = 20

import os
import numpy as np
import glob
import matplotlib.pyplot as plt

lp_dir = os.path.join(job_dir, f'pred_label_gt{gt}p{p}/seg')
gt_dir = os.path.join(data_dir, 'seg')

# list all seg files here, visualize 

# def vis_
fs = [x.split('/')[-1] for x in glob.glob(lp_dir+'/*.npy')]
fs.sort()

for file_name in fs:
    
    prop_data = np.load(os.path.join(lp_dir, file_name)).astype(np.uint8)
    gt_data = np.load(os.path.join(gt_dir, file_name)).astype(np.uint8)
    
    fig, axs = plt.subplots(1,2,figsize=(4,4))
    
    axs[0].axis('off')
    axs[0].set_title(file_name)
    axs[0].imshow(prop_data)

    axs[1].axis('off')
    axs[1].set_title(file_name)
    axs[1].imshow(gt_data)
    
    plt.show()

```
