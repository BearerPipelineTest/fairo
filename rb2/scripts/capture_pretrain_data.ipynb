```python
import numpy as np

# Record positions
sample_positions = []
for i in range(40):
    pos_str = input(f"EE position number {i+1}: ")
    pos = np.array([float(n) for n in pos_str.split(", ")])
    sample_positions.append(pos)
    
np.save("sample_positions.npy", np.array(sample_positions))
```

```python
import cv2

# Capture images
sample_positions = np.load("data/sample_positions.npy")

for i in range(7,8):
    for j in range(sample_positions.shape[0]):
        img_filename = f"data/pouring_train{i+1}_pos{j+1}.jpg"
        
        input(f"Press enter to capture train set {i+1} at position {j+1}")
        img = rs.get_images()[0]
        
        cv2.imwrite(img_filename, img)
```
